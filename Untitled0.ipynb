```python
import pandas as pd
import matplotlib.pyplot as plt

# создаем данные
data = {
    'hours_studied': [1, 2, 3, 4, 5, 6, 7, 8],
    'exam_score': [50, 55, 60, 65, 70, 75, 80, 90]
}

df = pd.DataFrame(data)

# смотрим данные
print(df)

# строим график
plt.scatter(df['hours_studied'], df['exam_score'])
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Study Time vs Exam Score')
plt.show()
```

Conclusion:
The analysis shows that students who study more hours tend to have higher exam scores.
This indicates a positive correlation between study time and performance.
