```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

# данные
texts = [
    "I love this product",
    "This is amazing",
    "I am very happy",
    "I hate this",
    "This is bad",
    "Very плохой продукт"
]

labels = [1, 1, 1, 0, 0, 0]  # 1 = positive, 0 = negative

# таблица
df = pd.DataFrame({'text': texts, 'sentiment': labels})

# превращаем текст в числа
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])

y = df['sentiment']

# модель
model = LogisticRegression()
model.fit(X, y)

# тест
test = ["I love it", "This is bad"]
test_vec = vectorizer.transform(test)

predictions = model.predict(test_vec)

print("Predictions:", predictions)
```

## Conclusion

The model successfully classifies text as positive or negative.
This demonstrates basic natural language processing and machine learning skills.
