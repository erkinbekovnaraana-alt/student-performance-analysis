```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# создаем данные
data = {
    'hours_studied': [1, 2, 3, 4, 5, 6, 7, 8],
    'passed': [0, 0, 0, 1, 1, 1, 1, 1]
}

# создаем таблицу
df = pd.DataFrame(data)

# разделяем данные
X = df[['hours_studied']]
y = df['passed']

# создаем модель
model = LogisticRegression()
model.fit(X, y)

# делаем предсказания
print("Prediction for 3 hours:", model.predict([[3]])[0])
print("Prediction for 6 hours:", model.predict([[6]])[0])

# график
plt.scatter(df['hours_studied'], df['passed'])
plt.xlabel('Hours Studied')
plt.ylabel('Passed (0 = No, 1 = Yes)')
plt.title('Pass/Fail Prediction')
plt.show()
```

## Conclusion

The model shows that students who study more hours are more likely to pass the exam.
There is a positive relationship between study time and passing probability.
